```python
from sympy import *
```

```python
a0, a1, a2 = symbols('a_0 a_1 a_2')

dt, l_dt = symbols('Δt Δt_l')
th, l_th, ll_th  = symbols('θ θ_l θ_ll')

t = 0
l_t = -dt
ll_t = l_t-l_dt
```

```python
A = Matrix([[1, t, t**2], [1, l_t, l_t**2], [1, ll_t, ll_t**2]])
B = Matrix([[th], [l_th], [ll_th]])

A.solve(B)[1].simplify()
```
